```python
# Chapter 11. Time Series

# Anything that is recorded repeatedly at many points in time forms a time series.
# Many time series are fixed frequency, which is to say that data points occur at regular intervals according to some rule
# Time Series can also be irregular without a fixed unit of time or offset between units.
# How to mark and refer to time series data depends on the application, that may have one of the following:

# Timestamps : Specific instants of time
# Fixed periods : Such as whole month of Junuary 2026, or the whole year of 2025.
# Intervals of time : Indicated by a start and end timestamp. Periods can be thought of as special cases of intervals.
# Experiment or elapsed time : Each timestamp is a measure of time relative to particular start time, starting from 0.
#                              (e.g., the diameter of a cookie baking each second since being placed in the oven)

# pandas also supports indexes based on timedeltas, which can be a useful way of representing experiment or elapsed time (not included in book though)

# pandas provides many built-in time series tools and algorithms.
# you can work with large time series, slicing, aggregate, and resample irregular and fixed frequency time series.

import numpy as np
import pandas as pd
```
