```python
# pitfalls with chained indexing

data = pd.DataFrame(np.arange(16).reshape((4,4)),
        index=["Ohio","Colorado","Utah","New York"],
        columns=["one","two","three","four"])
data
```

```python
data.loc[:,"one"] = 1
data
```

```python
data.iloc[2] = 5
data
```

```python
data.loc[data["four"]>5] = 3
data
```

```python
# data is unmodified as it suggests to use .loc instead
# This is called SettingWithCopyWarning

data.loc[data.three == 5]["three"] = 6
data
```

```python
# the fix is to rewrite the chained assignment to use a single loc operation
# and a good rule of thumb is to avoid chained indexing in general

data.loc[data.three == 5,"three"] = 6
data
```

```python
# Numpy ufuncs (element-wise array methods) also work with pandas objects
frame = pd.DataFrame(np.random.standard_normal((4,3)),
            columns=list("bde"),
            index=["Utah","Ohio","Texas","Oregon"])
frame
```

```python
# np.abs(): no negative filter, absolute value
np.abs(frame)
```

```python
# another frequent operation is applying a function on one-dimensional arrays to each column or row
# function f1 is invoked on each column in frame
# result is a Series having the columns of frame as index
def f1(x):
    return x.max() - x.min()
```

```python
frame.apply(f1)
```

```python
# if pass axis="columns", function will invoke once per row instaed
frame.apply(f1,axis="columns")
```

```python
# many of the most common array statistics (sum, mean) are DataFrame methods
# so using .apply is not necessary
def f2(x):
    return pd.Series([x.min(),x.max()],
    index=["min","max"])
frame.apply(f2)
```

```python
# element-wise functions
# computing a formatted string from each floating point value in frame
# using .applymap()

def my_format(x):
    return f"{x:.2f}"

# f"..." : formatting
# { } : placeholder, to take value from
# ":" : separator
# ".2f" : the specific instruction - .2 = keep exactly 2 decimal points; f: treat number as fixed-point (floating point) number
```

```python
# using .applymap() / applying ufunc
frame.applymap(my_format)
```

```python
# Series has a map method for applying element-wise function
# so, for DatraFrame applymap() is separately used
# Series.map()
# DataFrame.applymap()
frame["e"].map(my_format)
```

```python
# computing a histogram on multiple related columns in a DataFrame

data = pd.DataFrame({"Qu1":[1,3,4,3,4],
                     "Qu2":[2,3,1,2,3],
                     "Qu3":[1,5,2,4,4]})
data
```

```python
# computing value counts for a single column, .value_counts()
# .value_counts() returns a frequency of certain values in observations
data["Qu1"].value_counts().sort_index()
```

```python
# computing for all columns - pass pandas.valuecounts() to the DataFrame's apply method
# .apply() allows to apply a function along an axis of a DataFrame or Series
result = data.apply(pd.value_counts).fillna(0)
result

# the row labels in the result are the distinct values occurring in all of the columns
# the values are the respective counts of these values in each column
```
