```python
# 7.1 Handling Missing Data
```

```python
# for data with float64 dtype, pandas uses the floating-point value NaN (Not a Number) to represent misssing data
# this is called  asentienl value; when present, it indicates a missing (or null) value
import pandas as pd
import numpy as np

float_data = pd.Series([1.2,-3.5,np.nan,0])
float_data
```

```python
# .isna() gives a Boolean Series of result; null = TRUE
float_data.isna()
```

```python
# as alternative to 'NA', a built-in 'None' (case-sensitive) value is also used
string_data = pd.Series(["aardvark",np.nan,None,"avocado"])
string_data
```

```python
string_data.isna()
```

```python
float_data = pd.Series([1,2,None],dtype='float64')
float_data
```

```python
float_data.isna()
```

```python
# NA handling object methods
# .dropna() : filter axis labels based on whether values for eahc label have missing data
#          with varying thresholds for how much missing data to tolerate
# .fillna() : fill in missing data with some value or using an interpolation method such as "ffill" or "bfill"
# .isna() : return Boolean values indicating which values are missing/NA
# .notna() : negation of isna, returns True for non-NA values and False for NA values
```

```python
# Filtering Out Missing Data
```

```python
# using .dropna() in Series
data = pd.Series([1,np.nan,3.5,np.nan,7])
data
```

```python
data.dropna()
```

```python
# another way to filter out missing data in Series
data[data.notna()]
```

```python
# removing missing data in DataFrames - either drop rows or columns that are all NA, or only dropping rows or columns containing any NA
# .dropna() by default drops any row containing a missing value
data = pd.DataFrame([[1,6.5,3.],[1.,np.nan,np.nan],
                     [np.nan,np.nan,np.nan],[np.nan,6.5,3.]])
data
```

```python
data.dropna()
# drops all the rows containing any NA
```

```python
# passing how = "all" will drop only rows that are all NA
data.dropna(how="all")

# these functions don't modify the original object, just creates a new object
```

```python
# to drop columns containing NA, pass the argument axis = "columns"
data[4] = np.nan
data
```

```python
data.dropna(axis="columns",how="all")
```

```python
# only keeping rows contianing at most a certain number of missing observations
# indicate this with the thresh argument
df = pd.DataFrame(np.random.standard_normal((7,3)))
df
```

```python
df.iloc[:4,1] = np.nan
df.iloc[:2,2] = np.nan
df
```

```python
df.dropna()
```

```python
# using argument thresh
df.dropna(thresh=2)
# thresh = 2 means non-NaN threshold is 2 (at least 2 non-NaN values), if threshold not met, drop
# thresh = N means the row needs N "votes" to stay in the DataFrame
```

```python
# Filling In Missing Data
```

```python
# .fillna() : filling NaN with a constant
df.fillna(0)
```

```python
# using fillna() with a dictioinary, can use a different fill value for each column
df.fillna({1:0.5,2:0})
```

```python
# the same interpolation methods available for reindexing can be used with fillna()
df = pd.DataFrame(np.random.standard_normal((6,3)))
df
```

```python
df.iloc[2:,1] = np.nan
df
```

```python
df.iloc[4:,2] = np.nan
df
```

```python
df.fillna(method="ffill")
# "ffill" means forward fill; as in Excel "Duplicate Down"
# the NaN values are filled with the preceding non-NaN values, however, ffill stops when it hits a new non-NaN value
# opposite of "ffill" is "bfill" (back fill)
```

```python
df.fillna(method="ffill",limit=2)

# "limit = 2" means limit the fill to 2 rows
```

```python
# with .fillna() simple data imputation using median or mean statistics is available
data = pd.Series([1,np.nan,3.5,np.nan,7])
data
```

```python
data.fillna(data.mean())
# fill the NaN values with .mean()
```

```python
# .fillna() function arguments
# value : scalar value or dictionary-like object to use to fill missing values
# method : interpolation method - one of "bfill" or "ffill"; default is None
# axis : axis to fill on ("index" or "columns"); default is axis="index"
# limit : for ffill and bfill, maximum number of consecutive periods to fill
```

```python
# Removing Duplicates
```

```python
# removing duplicate rows
data = pd.DataFrame({"k1":["one","two"]*3 + ["two"],"k2":[1,1,2,3,3,4,4]})
data
```

```python
# DataFrame .duplicated() method returns a Boolean Series indicating whether each row is a duplicate oe nor
# duplicate means its column values are exactly equal to those in an earlier row
data.duplicated()
```

```python
# relatedly, drop_duplicates() returns a DataFrame with rows where the duplicated array is False filtered out
data.drop_duplicates()

# both methods, .duplicated() and .drop_duplicates() by default consider all of the columns;
# alternatively, can specify any subset of them to detect duplicates
```

```python
# having an additional column of values, and want to filter duplicates based only on the "k1" column
data["v1"] = range(7)
data
```

```python
# set column with subset argument
data.drop_duplicates(subset=["k1"])
```

```python
# .duplicated() and .drop_duplicates() by default keep the first observed value combination
# passing keep="last" argument will return the last one
data.drop_duplicates(["k1","k2"],keep="last")
```

```python
# Transformming Dta Using a Function or Mapping
# transformation based on the values in an array, Series, or column in a DataFrame
data = pd.DataFrame({"food":["bacon","pulled pork","bacon","pastrami","corned beef","bacon","pastrami","honey ham","nova lox"],
                     "ounces":[4,3,12,6,7.5,8,3,5,6]})
data
```

```python
# suppose we wanted to add a column indicating the type of animal for each food
meat_to_animal = {
    "bacon": "pig",
    "pulled pork": "pig",
    "pastrami": "cow",
    "corned beef": "cow",
    "honey ham": "pig",
    "nova lox": "salmon"
}

meat_to_animal
```

```python
# .map() method on a Series accepts a function or dictionary-like object containing a mapping
data["animal"] = data["food"].map(meat_to_animal)
data
```

```python
# could also have passed a function that does all the work
def get_animal(x):
    return meat_to_animal[x]

data["food"].map(get_animal)
```

```python
# using .map() is a convenient way to perform element-wise transformatioins and other data cleaning-related operations
```

```python
# Replacing Values
# .map() can be used to modify a subset of values in an object,
# but .replace() provides a simpler and more flexible way
data = pd.Series([1.,-999.,2.,-999.,-1000.,3.])
data
```

```python
# using "-999" values as sentinel values for missing data - replacing those with NA values
data.replace(-999,np.nan)
```

```python
# replacing multiple values at once - passing a list with substitute values
data.replace([-999,-1000],np.nan)
```

```python
# replacing different values for each, pass a list of substitutes
data.replace([-999,-1000],[np.nan,0])
```

```python
# passing argument can also be a dictionary
data.replace({-999:np.nan,-1000:0})
```

```python
# renaming axis indexes
# axis labels can be similarly transformed by a function or mapping of some form to produce new, differently labeled objects
# modifying the axes in place without creating a new data structure
data = pd.DataFrame(np.arange(12).reshape((3,4)),
        index = ["ohio","Colorado","New York"],
        columns = ["one","two","three","four"])
data
```

```python
# .map() method for axis indexes, just like in Series
def transform(x):
    return x[:4].upper()

data.index.map(transform)
```

```python
# can assign to the index attribute, modifying the DataFrame in place
data.index = data.index.map(transform)
data
```

```python
# if wanting to create a transformed version without modifying the origina, .renam() can be used
data.rename(index=str.title,columns=str.upper)
```

```python
# .rename() can be used in conjuction with a dictionary-like object
data.rename(index={"OHIO":"INDIANA"},
           columns={"three":"peekaboo"})
```

```python
# Discretization and Binning
# data about a group of people in a study, gropuping them into discreet age buckets
ages = [20,22,25,27,21,23,37,31,61,45,41,32]
ages
```

```python
# bins = 18 to 25, 26 to 35, 36 to 60, and 61 and older
# use pd.cut()
bins = [18,25,35,60,100]
age_categories = pd.cute(ages,bins)
age_categories
```

```python
bins = [18,25,35,60,100]
age_categories = pd.cut(ages,bins)
age_categories
# each bin is identified by a special interval value type containing lower and upper limit of each bin
```

```python
# data codified in numberical integers (the "pointer" to the category)
age_categories.codes
```

```python
# index of unique labels created; telling what the buckets are called, the "reference table"
age_categories.categories
```

```python
age_categories.categories[0]
```

```python
# pd.value_counts() are the bin counts for the result of pd.cut()
pd.value_counts(age_categories)
```

```python
# in the string representation of an interval, a parenthesis means that the side is open (exclusive)
# square bracket means it's closed (inclusive)
# can change which side is closed by passing "right = False"
pd.cut(ages,bins,right=False)
```

```python
# can overrid the default interval-based bin labeling, by passing a list or arroy to the labels
group_names = ["Youth","YoungAdult","MiddleAged","Senior"]
pd.cut(ages,bins,labels=group_names)
```

```python
# passing an integer number of bins to pd.cut() instead of explicit bin edges,
# it will compute equal-length bins based on min and max values
data = np.random.uniform(size=20)
data
```

```python
pd.cut(data,4,precision=2)
```

```python
# pd.qcut() bins the data based on sample quantiles
# it will obtain roughly equally sized bins (slicing in quantiles)
data = np.random.standard_normal(1000)
data
```

```python
quartiles = pd.qcut(data,4,precision=2)
quartiles
```

```python
pd.value_counts(quartiles)
```

```python
# similar to pd.cut(), can pass own quantiles (between 0 and 1, inclusive)
pd.qcut(data,[0,0.1,0.5,0.9,1]).value_counts()
```

```python
# Permutation and Random Sampling
# permuting (randomly reordering) the rows in a DataFrame
# np.random.permuitation() with length of the axis wanting to permute prodcues an array of integers of new ordering
df = pd.DataFrame(np.arange(5*7).reshape((5,7)))
df
```

```python
# Computing Indicator / Dummy Variables
# common statistical modeling or machine learning application
# converting a categorical variable into a dummy or indicator matrix
# if column in a DataFrame has k distinct values, derive a matrix or DataFrame with k column containing all 1s and 0s
# pd.get_dummires() 
df = pd.DataFrame({"key":["b","b","a","c","a","b"],
                  "data1":range(6)})
df
```

```python
pd.get_dummies(df["key"])
```

```python
# in some cases, adding a prefix to the columns in the indicator DataFrame could be helpful
# prefix can e merged with the other data
# pd.get_dummies() has a prefix argument
dummies = pd.get_dummies(df["key"],prefix="shit")
dummies
```

```python
df_with_dummy = df[["data1"]].join(dummies)
df_with_dummy
```

```python
# if row in a DataFrame belongs to multiple categories, use different approach to create dummy varaiables
mnames = ["movie_id","title","genres"]
```

```python
movies = pd.read_table("datasets/movielens/movies.dat",sep="::",
                      header=None,names=mnames,engine="python")
movies[:10]
```

```python
dummies.iloc[:10,:6]
```

```python
# as before, can combine this with movies while adding a "Genre_" to the column names in dummies DataFrame
# .add_prefix()
movies_windic = movies.join(dummies.add_prefix("Genre_"))
movies_windic
```
