```python
# only keeping rows contianing at most a certain number of missing observations
# indicate this with the thresh argument
df = pd.DataFrame(np.random.standard_normal((7,3)))
df
```

```python
df.iloc[:4,1] = np.nan
df.iloc[:2,2] = np.nan
df
```

```python
df.dropna()
```

```python
# Regular Expressions
# regular expressions provide a felxible way to search or match string patterns in text
# a single expression, commonly called regex, is a string formed according to the regular expression language

# re module functions fall into three categories:
# pattern matching, substitution, and splitting
# splitting a variable number of whitespace characters - tabs, spaces, and newlines
# the regex describing one or more whitespace characters is \s+:

import re
text = "foo  bar\tbaz \tqux"
text
```

```python
re.split(r"\s+",text)
```

```python
# always use r for regex pattern; r stands for "Raw"
# \s means "any whitespace"; + means "one or more"
# and import re! so, re.split(r"\s+", text) is like Excel's TRIM()
```

```python
# when calling re.split(r"\s+", text), the regex is first compiled, and then its .split() is called on the passed text
# can compile the regex yourself with re.compile(); this forms a reusable regex object
regex = re.compile(r"\s+")
```

```python
regex
```

```python
regex.split(text)
```

```python
# instead, if wanted to get a list of all patterns matching the regex, can use .findall()
regex.findall(text)
```

```python
# creating a regex object with re.compile() is highly recommended, for saving CPU cycles
```

```python
# match and search are closly related to .findall()
# while .findall() returns all matches in a string, .search() returnls only the first match
# .match() only matches at the beginning of the string
# email addresses as example
text = """Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com"""
pattern = r"[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}"

# [A-Z0-9._%+-] : Find those patterns, any character in that is allowed
# + : "one or more"
# @ : find "@"
# [A-Z0-9.-] : A-Z, 0-9, ., - are allowed
# + = "one or more"; \. : find "."
# [A-Z]{2,4} : find A-Z but between 2 and 4 charecters long (e.g. .com, .de, .info)

regex = re.compile(pattern,flags=re.IGNORECASE)
# compile (contain) from the object "pattern" into object "regex", ignore cases
```

```python
# using .findall()
regex.findall(text)
```

```python
# .search() returns a special match object for the first email address in the text
# .match() object can only tell us the start and end position of the pattern in the string
m = regex.search(text)
m
```

```python
text[m.start():m.end()]
```

```python
# regex.match returns None, as it will match only if the pattern occurs at the start of the string
print(regex.match(text))
```

```python
# .sub() will return a new string with occurrences of the pattern replaced by a new string
print(regex.sub("REDACTED",text))

# .sub() functions like Excel's "Find and replace"
```

```python
# suppose you want to segment each addres into 3 components - user name, domain name, and domain suffix
pattern = r"([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})"
regex = re.compile(pattern,flags=re.IGNORECASE)
```

```python
# returning a tuple of pattern components with .groups() method
m = regex.match("wesm@bright.net")
m.groups()
```

```python
# findall() returns a list of tuples when pattern has groups
regex.findall(text)
```

```python
# .sub() also has access to .groups() in each .match() using special symbols; \1 \2
# \1 corresponds to the first matched group, \2 corresponds to the second, and so forth
print(regex.sub(r"Username:\1,Domain:\2,Suffix:\3",text))
```

```python
# more regular expressions
# .findall() : Return all noneoverlapping matching patterns in a string as a list
# .finditer : Like .findall() but returns an iterator
# .match() : Match pattern at start of the string and optionally segment pattern components into groups;
#            if the pattern matches, return a match object, and otherwise None
# .search() : Scan string for match to pattern, returning a match object if so;
#             unlike .match(), the match can be anywhere in the string
# .split() : Break string into pieces at each occurrence of pattern
# .sub(), .subn() : Replace all (sub) or first n occurrences (subn) of pattern in string with replacement expression;
#                   use symbols \1, \2, ... to refer to match group elements in the replacement string 
```
